#  **Data Pre-Processing**

```python
TrainData.head
```

```python
TrainData.drop(['Name','Cabin','Ticket'],inplace=True, axis= 'columns')
TestData.drop(['Name','Cabin','Ticket'],inplace=True, axis= 'columns')
```

```python
data = [TrainData, TestData] 

for dataset in data:
      dataset['Embarked'] = dataset['Embarked'].fillna('S')
        
        
        
```

```python
TrainData.isnull().sum()
```

```python
numericSex = {"male": 0, "female": 1}
numericEmbarked = {"S": 0, "C": 1, "Q":2}

data = [TrainData, TestData] 

for dataset in data:
    dataset['Sex'] = dataset['Sex'].map(numericSex)
    dataset['Embarked'] = dataset['Embarked'].map(numericEmbarked)
```

```python
TestData.head(10)
```

```python
TrainData.head(10)
```

```python
TrainData["Age"].fillna(TrainData["Age"].median(), inplace=True)
TrainData.isnull().sum()
```

```python
TestData["Age"].fillna(TestData["Age"].median(), inplace=True)
TestData.isnull().sum()
```

```python
TestData["Fare"].fillna(TestData["Fare"].median(), inplace=True)
TestData.isnull().sum()
```

```python
TestData["Fare"].fillna(TestData["Fare"].median(), inplace=True)
TestData.isnull().sum()
```

```python
TrainData.head(10)
```

```python
TestData.head(10)
```

```python
featureDrop = ['SibSp','Parch']
train = TrainData.drop(featureDrop, axis=1)
test = TestData.drop(featureDrop, axis=1)
```

```python
train_data = train.drop('Survived', axis=1)
target = train['Survived']

train_data.shape, target.shape
```

```python
train.head(10)
```

```python
test.head(10)
```

# **Model Training**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
TitanicModel = RandomForestClassifier(random_state = 1) #defining Random Classifier model
TitanicModel.fit(train_data, target) #training model
test_predictions = TitanicModel.predict(test) #prediction 
```

```python
output = pd.DataFrame({'PassengerId': TestData.PassengerId,
                       'Survived': test_predictions})
out = pd.DataFrame(data = output)
out.to_csv('submissionRF.csv', index=False)
```

# **Model Evaluation**

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
train_X, val_X, train_y, val_y = train_test_split(train_data, target,test_size=0.3, random_state=1)
TitanicModel.fit(train_X, train_y)
Titanic = TitanicModel.predict(val_X)
```

```python
kfold = KFold(n_splits=10, random_state = 1, shuffle = True) 
res=cross_val_score(TitanicModel,train_X,train_y,cv= kfold,scoring='accuracy')
print(res)
print('Cross validated score is:',round(res.mean()*100,2))
```

```python
accuracy_score(val_y,Titanic)
```
